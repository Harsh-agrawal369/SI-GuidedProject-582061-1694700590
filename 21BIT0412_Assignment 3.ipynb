```python
#Performing Encoding

#As we observe there are 3 categorical column
#we will encode them using label encoder

from sklearn.preprocessing import LabelEncoder

```

```python
le = LabelEncoder()

k= df.species
df.species=le.fit_transform(k)
m=df.island
df.island=le.fit_transform(m)
l= df.sex
df.sex= le.fit_transform(l)
```
